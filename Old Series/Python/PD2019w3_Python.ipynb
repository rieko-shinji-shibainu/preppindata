```python
import pandas as pd
```

```python
# インポートファイルを読み込む
scaffold = pd.read_csv('/content/Scaffold.csv')
scaffold.head(10)
```

```python
customer = pd.read_csv('/content/Preppin’ Data - Week Three.csv')
customer.head(10)
```

```python
# scaffold不要列の削除
scaffold_dropcol = scaffold.drop(['Unnamed: 1'], axis=1)
scaffold_dropcol.head(10)
```

```python
# scaffoldのnullを削除
scaffold_dropcol = scaffold_dropcol.dropna()
scaffold_dropcol.head(10)
```

```python
# scaffoldのLengthを整数に
scaffold_dropcol['Length'] = scaffold_dropcol['Length'].astype(int)
scaffold_dropcol.head(10)
```

```python
# customerの不要な列を削除
customer_dropcol = customer.drop(['Unnamed: 4'], axis=1)
customer_dropcol.head(10)
```

```python
# customerのNaNを削除
customer_dropna = customer_dropcol.dropna()
customer_dropna.head(10)
```

```python
# costomerのContranct Lengthを整数に変換
customer_dropna['Contract Length'] = customer_dropna['Contract Length'].astype(int)
customer_dropna.head(10)
```

```python
# customerとscaffoldを結合
# 不等号結合のためにpyjanitorのconditional_joinを使用

!pip install pyjanitor
import janitor
```

```python
joined = customer_dropna.conditional_join(
    scaffold_dropcol,
    ('Contract Length','Length','>=')
)
joined.head(10)
```

```python
# Start Dateの型を日付に変更
joined['Start Date'] = pd.to_datetime(joined['Start Date'])
joined.dtypes
```

```python
# Start DateをLength分加算してPayment Dateを作成
# 各行に対してMonthEndを適用
joined['Payment Date'] = joined.apply(lambda row: row['Start Date'] + pd.DateOffset(months=row['Length']-1), axis=1)
# lamda:無名関数
# apply:各要素に対して1行ずつ関数を適用
joined.head(10)
```

```python
# Monthly Costを整数に変換
joined['Monthly Cost'] = joined['Monthly Cost'].astype(int)
joined.head(10)
```

```python
# 正解データを読み込む
solution = pd.read_csv('/content/Contract Revenue Data.csv')
solution.head(10)
```

```python
# 列を並び替えて不要列を削除
new_order = ['Payment Date', 'Name', 'Monthly Cost', 'Contract Length', 'Start Date']
output = joined.reindex(columns=new_order)
output.head(10)
```

```python
# solutionとoutputの型確認
solution.dtypes
```

```python
# solutionの日付列を日付型に変更
solution['Payment Date'] = pd.to_datetime(solution['Payment Date'])
solution['Start Date'] = pd.to_datetime(solution['Start Date'])
solution.dtypes
```

```python
## outputとsolutionをマージさせる
a_merge = pd.merge(output,solution,how="outer",on=['Payment Date', 'Name', 'Monthly Cost', 'Contract Length', 'Start Date'],indicator=True)

# outputにしかない行
# left_only = a_merge[a_merge['_merge'] == 'left_only']
# left_only

# solutionにしかない行
right_only = a_merge[a_merge['_merge'] == 'right_only']
right_only
```
