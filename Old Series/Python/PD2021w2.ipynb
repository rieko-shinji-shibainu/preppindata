```python
# pandasをインポート
import pandas as pd
```

```python
# csvを読み込み
df = pd.read_csv('/content/PD 2021 Wk 2 Input - Bike Model Sales.csv')
```

```python
# データの確認
df.head()
```

```python
# 各列のデータ型の確認
df.dtypes
```

```python
# Order DateとShipping Dateがイギリス式の日付になっているので、yyyy/mm/ddに直す
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')
df['Shipping Date'] = pd.to_datetime(df['Shipping Date'], format='%d/%m/%Y')
df.head()
```

```python
# 各列のデータ型の確認
df.dtypes
```

```python
# 正規表現がうまくいかないので文字列を抽出する関数を作る
import string
def extract_chars(text):
    result = ""
    for char in text:
        if char.isalpha():
            result += char
    return result
```

```python
# Model列から文字のみを抽出
df['Model'] = df['Model'].apply(extract_chars)
df
```

```python
# Model列の値を確かめる
df['Model'].unique()
```

```python
# Value per BikeとQuantityをかけたOrder Value列を作成
df['Order Value'] = df['Value per Bike'] * df['Quantity']
df.head()
```

```python
# Model列の列名をBrandに変更
rename = df.rename(columns={'Model': 'Brand'})
rename.head()
```

```python
# Order DateとShipping Dateの差の日数を表すDays to Ship列を作成
rename['Days to Ship'] = rename['Shipping Date'] - rename['Order Date']
rename.head()
```

```python
# Days to Ship列の値を確認
rename['Days to Ship'].unique()
```

```python
# Days to ship列を整数に型変換
rename['Days to Ship'] = rename['Days to Ship'].dt.days
rename.head()
```

```python
# Days to ship列の型を確認
rename['Days to Ship'].dtypes
```

## Storeごとのデータ（出力1）を作成

```python
# Brandごと、StoreごとにDays to Shipの平均、Order Valueの合計、Quantityの合計で集計する
ByStore = rename.groupby(['Brand', 'Store']).agg({'Days to Ship': 'mean', 'Order Value': 'sum', 'Quantity': 'sum'})
ByStore.head()
```

```python
# Days to ship列を小数点第1位までの表示に丸める
ByStore['Avg Days to Ship'] = ByStore['Days to Ship'].round(1)
ByStore.head()
```

```python
# Days to Ship列を削除
ColumnRemove = ByStore.drop('Days to Ship', axis=1)
ColumnRemove.head()
```

```python
# Order Value→Total Order Value, Quantity→Total Quantity Soldに変更
ColumnRename = ColumnRemove.rename(columns={'Order Value': 'Total Order Value', 'Quantity': 'Total Quantity Sold'})
ColumnRename.head()
```

→出力1が完成

```python
# Brandごと、Bike TypeごとにValue per Bikeの平均、Order Valueの合計、Quantityの合計で集計する
ByBrand = rename.groupby(['Brand', 'Bike Type']).agg({'Value per Bike': 'mean', 'Order Value': 'sum', 'Quantity': 'sum'})
ByBrand.head()
```

```python
# Quantity→Quantity Soldに変更
BrandColumnRename = ByBrand.rename(columns={'Quantity': 'Quantity Sold'})
BrandColumnRename.head()
```

```python
# Order Value列を小数点第1位までの表示に丸める
BrandColumnRename['Order Value'] = BrandColumnRename['Order Value'].round(1)
BrandColumnRename.head()
```

```python
# Value per Bike列を小数点第1位までの表示に丸めて新しい列を作成
BrandColumnRename['Avg Bike Value Sold per Brand, Type'] = BrandColumnRename['Value per Bike'].round(1)
BrandColumnRename.head()
```

```python
# Value per Bike列を削除
BrandColumnRemove = BrandColumnRename.drop('Value per Bike', axis=1)
BrandColumnRemove.head()
```

```python
# 列の順番を入れ替え
BrandColumnOrdered = BrandColumnRemove.reindex(columns=['Quantity Sold','Order Value','Avg Bike Value Sold per Brand, Type'])
BrandColumnOrdered.head()
```

→出力2の完成

### 答え合わせ

#### 出力1答え合わせ

```python
# 出力1回答のcsv読み込み（Storeごとデータ）
answer1 = pd.read_csv('/content/PD 2021 Wk 2 Brand Store.csv')
answer1.head()
```

```python
## answer1とColumnRenameをマージさせる
a_merge = pd.merge(ColumnRename,answer1,how="outer",on=['Brand', 'Store', 'Total Order Value', 'Total Quantity Sold', 'Avg Days to Ship'],indicator=True)

# outputにしかない行
# left_only = a_merge[a_merge['_merge'] == 'left_only']
# left_only.head()

# solutionにしかない行
right_only = a_merge[a_merge['_merge'] == 'right_only']
right_only
```

→left_only, right_only共に0件なので正解

#### 出力2答え合わせ

```python
# 出力2回答のcsv読み込み（Storeごとデータ）
answer2 = pd.read_csv('/content/PD 2021 Wk 2 Brand Type.csv')
answer2.head()
```

```python
## answer2とBrandColumnOrderedをマージさせる
a2_merge = pd.merge(BrandColumnOrdered,answer2,how="outer",on=['Brand', 'Bike Type', 'Quantity Sold','Order Value', 'Avg Bike Value Sold per Brand, Type'],indicator=True)

# outputにしかない行
# left_only2 = a2_merge[a2_merge['_merge'] == 'left_only']
# left_only2.head()

# solutionにしかない行
right_only2 = a2_merge[a2_merge['_merge'] == 'right_only']
right_only2
```

→left_only, right_only共に0件なので正解
